```python
import pickle as pkl
import matplotlib.pyplot as plt
```

```python
DATASET = 'PBMC1'
TUNING = 'celltypist'
TOOLS = ['seurat', 'scanpy']#, 'monocle', 'scvi-tools', 'COTAN']
STEP = 100
START = 100
N_MARKERS = 500
```

```python
with open(f"./Data/{DATASET}/rf_{TUNING}_f1_increasing.pkl", 'rb') as fp:
	f1 = pkl.load(fp)
```

```python
cluster_names_to_plot = ['Classical monocytes', 'Naive B cells']
for cluster in cluster_names_to_plot:
    fig, ax = plt.subplots(1, len(TOOLS), figsize=(20,5))
    for i, tool in enumerate(TOOLS):
        ax[i].plot([i for i in range(START, N_MARKERS+1, STEP)], f1[tool][cluster], '-o', label=tool)
        ax[i].set_title(tool)
        ax[i].set_xlabel('Number of markers')
        ax[i].set_ylabel('F1 score')
    fig.suptitle(cluster, fontweight='bold')
```
